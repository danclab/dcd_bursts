```python
import pandas as pd
import numpy as np
import pyreadstat
```

```python
preproc_df=pd.read_csv('/home/bonaiuto/dcd_ross/data/derivatives/NEARICA_behav/NEARICA_preprocessing_report_11-05-2025.csv')
df, meta = pyreadstat.read_sav('/home/bonaiuto/dcd_ross/CoMB_Dataset.sav')
```

```python
preproc_df['num_lof_bad_channels'] = preproc_df['lof_bad_channels'].apply(
    lambda x: 0 if pd.isna(x) else len(str(x).split(','))
)
preproc_df['num_ICs_removed'] = preproc_df['ICs_removed'].apply(
    lambda x: 0 if pd.isna(x) else len(str(x).split('  '))
)
preproc_df['total_task_epochs_removed']=preproc_df['total_task_epochs_before_artifact_rejection']-preproc_df['total_task_epochs_after_artifact_rejection']
```

```python
# Merge the two dataframes on subject ID
merged_df = preproc_df.merge(df[['SubjectID', 'Group']], left_on='subject', right_on='SubjectID', how='inner')

# Compute and print the stats for each group
group_stats = merged_df.groupby('Group')['num_lof_bad_channels'].agg(['min', 'max', 'mean', 'std']).reset_index()

# Rename the groups for clarity
group_stats['Group'] = group_stats['Group'].map({0: 'Control', 1: 'DCD'})

# Print the results
for _, row in group_stats.iterrows():
    print(f"{row['Group']}: {row['min']} - {row['max']}, M = {row['mean']:.2f}, SD = {row['std']:.2f}")
```

```python
# Compute and print the stats for each group
group_stats = merged_df.groupby('Group')['total_task_epochs_removed'].agg(['min', 'max', 'mean', 'std']).reset_index()

# Rename the groups for clarity
group_stats['Group'] = group_stats['Group'].map({0: 'Control', 1: 'DCD'})

# Print the results
for _, row in group_stats.iterrows():
    print(f"{row['Group']}: {row['min']} - {row['max']}, M = {row['mean']:.2f}, SD = {row['std']:.2f}")
```
